### Group By Functions:
1.split, 2.Apply Functions, 3.Combine Results

```python
df['Country'].value_counts()
```

```python
country_grp=df.groupby(['Country'])
```

```python
country_grp.get_group('India')
```

```python
filt = df['Country']  == 'India'
df.loc[filt]['UndergradMajor'].value_counts()
```

```python
country_grp['UndergradMajor'].value_counts().loc['India']
```

```python
country_grp['SalriesUSD'].median()
```

```python
country_grp['SalriesUSD'].median().loc['India']
```

```python
# for mean and median

country_grp[['SalriesUSD',]].agg(['mean','median']).loc[['India', 'United States']]
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

### Converting into percentage

```python
country_respondents = df['Country'].value_counts()
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

### Remaning the colomns

```python
python_df.rename(columns={'Country':'NumRespondents', 'LanguageWorkedWith':'NumKnowsPython'})
```
